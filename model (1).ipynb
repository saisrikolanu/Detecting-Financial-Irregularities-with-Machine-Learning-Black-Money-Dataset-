# **Data Cleaning Steps**

#1.Check for Missing Data

```python
missing_data = data.isnull().sum()
missing_data
```

## #1a.Handle Missing Data
### Dropping rows with critical missing info like Transaction ID

```python
#1a.Handle Missing Data
# Dropping rows with critical missing info like Transaction ID
data = data.dropna(subset=['Transaction ID', 'Country', 'Amount (USD)', 'Transaction Type','Date of Transaction', 'Person Involved', 'Industry','Destination Country', 'Reported by Authority', 'Source of Money',
'Money Laundering Risk Score', 'Shell Companies Involved', 'Financial Institution', 'Tax Haven Country'])
data
```

## 1b.Fill missing amounts with median value

---



```python
data['Amount (USD)'] = data['Amount (USD)'].fillna(data['Amount (USD)'].median())
data
```

# 2.Remove Duplicates

```python
data = data.drop_duplicates()
data
```

# 3. Standardize Date Formats

```python
data['Date of Transaction'] = pd.to_datetime(data['Date of Transaction'])
data
```

# 4.Check Date and Time Correlation

```python

data = data[data['Date of Transaction'] <= pd.Timestamp.today()]
data
```

# 5.Remove Unwanted data

```python
data=data.drop(['Tax Haven Country'], axis=1)
data
```

# 6.Check for outliers using BoxPlot

```python

plt.figure(figsize=(8, 6))
sns.boxplot(data=data, x='Amount (USD)')
plt.title('Boxplot for Amount (USD) - Outlier Detection')
plt.xlabel('Amount (USD)')
plt.show()
```

#7 .For high-risk transactions(>7)

```python

data['High Risk'] = data['Money Laundering Risk Score'] > 7
data
```

#8.Categorize 'Transaction Type'

```python

data['Transaction Type'] = data['Transaction Type'].str.lower().replace({'dep': 'deposit', 'wd': 'withdrawal'})
data
```

#9.Label Encoding

```python
from sklearn.preprocessing import LabelEncoder

categorical_columns = ['Transaction ID','Source of Money','Transaction Type', 'Person Involved', 'Industry', 'Country', 'Destination Country','Financial Institution','High Risk','Date of Transaction']

for column in categorical_columns:
  data[column] = data[column].astype(str)
  le = LabelEncoder()
  data[column] = le.fit_transform(data[column])

data.head()
```

#10.Sorting the data with Country Name

```python

data = data.sort_values(by='Country', ascending=True)
data
```

#11.Remove non-ASCII characters from Person Involved

```python

data['Person Involved'] = data['Person Involved'].astype(str)
data['Person Involved'] = data['Person Involved'].str.encode('ascii', 'ignore').str.decode('ascii')
data['Person Involved']
```

##**MODEL** **BUILDING**

SPLITTING DATA FOR TRAINING AND TESTING

```python
from sklearn.model_selection import train_test_split

X = data.drop(['High Risk'],axis=1)
y = data['High Risk']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)
```

**1. LOGISTIC REGRESSION**

modeling

```python
from sklearn.linear_model import LogisticRegression
log_reg= LogisticRegression(penalty='l2',
                            C=1.0, random_state=20,
                            solver='lbfgs', max_iter=100)

log_reg.fit(X_train,y_train)
```

Prediction

```python
log_pred=log_reg.predict(X_test)
```

Accuracy

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

ac_log=accuracy_score(y_test,log_pred)
print('Accuracy:',ac_log)
```

Confusion Matrix

```python
log_cm=confusion_matrix(y_test,log_pred)
log_cm
```

```python
plt.figure(figsize=(4, 4))
sns.heatmap(log_cm, annot=True, fmt='d')
plt.title('Confusion Matrix Heatmap')
plt.xlabel('Predicted value',fontsize=12)
plt.ylabel('actual value',fontsize=14)
```

Classification Report

```python
from sklearn.metrics import classification_report
log_rep=classification_report(y_test,log_pred, output_dict=True)
log_rep=pd.DataFrame(log_rep).transpose()
log_rep
```
